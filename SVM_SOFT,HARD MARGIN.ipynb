```python
from sklearn.svm import SVC

classifier = SVC(kernel = 'linear')

training_points = [[1,2], [1,5], [2,2], [7,5], [9,4], [8,2], [4,4], [6,3], [8,6], [5,3], [6,5]]
labels = [1,1,1,0,0,0,1,0,0,1,0]

classifier.fit(training_points, labels)
```

```python
print(classifier.predict([[3,2]]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.scatter([p[0] for p in training_points], [p[1] for p in training_points],
            c=labels, s=100, edgecolors='k', cmap=plt.cm.coolwarm)
plt.xlabel("X1")
plt.ylabel("X2")
plt.yticks(np.arange(0, 8, 1))
plt.title("example")
plt.show()
```

```python
def plot_svm(C_value, title):
    
    classifier = SVC(kernel='linear', C=C_value)
    classifier.fit(training_points, labels)
    
    x_min, x_max = 0, 10
    y_min, y_max = 0, 7
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    Z = classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.scatter([p[0] for p in training_points], [p[1] for p in training_points],
                c=labels, s=100, edgecolors='k', cmap=plt.cm.coolwarm)
    plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='k')
    plt.title(title)
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.show()
```

```python
plot_svm(C_value=10000, title="Hard Margin")

plot_svm(C_value=0.1, title="Soft Margin")
```
